```python
import pandas as pd
```

```python
df_busisness = pd.read_parquet('./data_new/Busisness.parquet')
df_reviews_2015_2018 = pd.read_parquet('./data/reviews2015-2018.parquet')
df_reviews_2019_2022 = pd.read_parquet('./data/reviews2019-2022.parquet')
```

```python
review_ids = df_busisness['business_id'].tolist()
reviews_2015 = df_reviews_2015_2018[df_reviews_2015_2018['business_id'].isin(review_ids)]
reviews_2022 = df_reviews_2019_2022[df_reviews_2019_2022['business_id'].isin(review_ids)]
```

```python
df_combined_reviews = pd.concat([reviews_2015, reviews_2022], ignore_index=True)
```
