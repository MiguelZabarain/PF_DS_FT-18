```python
import pandas as pd
import numpy as np
import nltk
import pyarrow as pa
import pyarrow.parquet as pq

from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
df_reviews = pd.read_parquet('gs://datos-machine-learning/users.parquet')
```

```python
df_reviews
```

```python
comentarios = df_reviews['text']
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
for i, comentario in enumerate(comentarios):
    scores = analyzer.polarity_scores(comentario)
    # Asigna 1 si el sentimiento es positivo, 0 si es negativo o neutral
    df_reviews.loc[i, 'sentimiento'] = 1 if scores['compound'] > 0 else 0
```

```python
df_reviews
```

```python
df_reviews['sentimiento'] = df_reviews['sentimiento'].astype(int)
```

```python
df_reviews = df_reviews.drop('text', axis=1)
```

```python
df_reviews = df_reviews.drop('user_id', axis=1)
```

```python
df_reviews = df_reviews.drop('compliment_count', axis=1)
```

```python
df_reviews = df_reviews.drop('date',axis=1)
```
