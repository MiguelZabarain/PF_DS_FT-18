```python
import pyarrow as pa
import pyarrow.parquet as pq
import pandas as pd
import numpy as np
```

```python
df_busisnes = pd.read_parquet('gs://data_henry_proyecto_final/busisnes-clean.parquet')
```

```python
df_checkin = pd.read_parquet('gs://data_henry_proyecto_final/checkin-clean.parquet')
```

```python
df_reviews = pd.read_parquet('gs://data_henry_proyecto_final/reviews-clean.parquet')
```

```python
df_tip = pd.read_parquet('gs://data_henry_proyecto_final/tip-clean.parquet')
```

```python
df_user = pd.read_parquet('gs://data_henry_proyecto_final/user-clean.parquet')
```

```python
df_reviews
```

```python
df_reviews = df_reviews.merge(df_user, on='user_id', how='outer')
```

```python
df_reviews
```

```python
df_reviews = df_reviews.merge(df_tip, on=['business_id','user_id'], how='outer')
```

```python
df_reviews = df_reviews.merge(df_checkin, on='business_id', how='outer')
```

```python
df_reviews = df_reviews.merge(df_busisnes, on='business_id', how='outer')
```

```python
df_reviews
```

```python
import dask.dataframe as dd
```

```python
table = pa.Table.from_pandas(df_reviews)
schema = table.schema

with pq.ParquetWriter('gs://data_henry_proyecto_final/reviews-clean.parquet', schema) as writer:
    writer.write_table(table)
```
